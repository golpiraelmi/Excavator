```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

# Read the text and set the stopwords (a, an, the, have, etc.)
text = open('TheHappyPrince.txt').read()
stopwords = set(STOPWORDS)

# Creating a matrix of colours from the image
coloring = np.array(Image.open("EXCAVATOR.jpg")) 

# creating coloring from image
image_colors = ImageColorGenerator(coloring)

# generate word cloud from text
wc = WordCloud(background_color="white", max_words=500, mask=coloring,
               stopwords=stopwords, max_font_size=50, random_state=0)
wc.generate(text)

# Create the graph
fig = plt.figure(figsize = (14,7))
plt.imshow(wc.recolor(color_func=image_colors, random_state=0),
           interpolation="bilinear")
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis('off')
plt.savefig('Excavator.jpeg', dpi=600)
plt.show()
```
